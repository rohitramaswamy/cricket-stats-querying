```python

# Initialize current_df correctly
current_df = pd.DataFrame()
# Function to remove text within parentheses
stop_year = 1999
for i in range(year - stop_year):

    url = f'https://www.worldfootball.net/goalgetter/eng-premier-league-{year-1}-{year}/'
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Check if the request was successful

        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all tables in the page
        tables = soup.find_all('table')

        # Convert the first table to a pandas DataFrame
        if tables:
            current_df = pd.DataFrame(pd.read_html(str(tables[1]))[0])[['Player', 'Goals (Penalty)']]
            current_df.rename(columns={'Goals (Penalty)': f'{year - 1}'}, inplace=True)

            current_df[f'{year - 1}'] = current_df[f'{year - 1}'].str.replace(r'\s*\([^)]*\)', '', regex=True).astype(int)
        else:
            print("No tables found on the page.")
    except requests.HTTPError as e:
        print(f"HTTP error occurred: {e.response.status_code} - {e.response.reason}")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

    merged_df = pd.merge(df[['Player',f'{year}']], current_df, on='Player', how='left')

    merged_df[f'{year - 1}'].fillna(0, inplace=True)
    # Subtract the goals
    merged_df['Goals'] = merged_df[f'{year}'] - merged_df[f'{year - 1}']
    


    # Select only the relevant columns
    df[f'{year - 1}'] = merged_df['Goals']

    year = year - 1

df_reversed = df[df.columns[::-1]]

print(df_reversed)

```

```python
excel_file = "new.xlsx" 
 
# Write DataFrame to Excel 
df_reversed.to_excel(excel_file, index=False) 
```

```python
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({
    'player': ['player1', 'player2', 'player3'],
    'goals': [10, 20, 30]
})

df2 = pd.DataFrame({
    'player': ['player1', 'player2', 'player4','player5'],
    'goals': [5, 15, 25, 30]
})

# Left merge to keep all players from df1
merged_df = pd.merge(df1, df2, on='player', how='left', suffixes=('_df1', '_df2'))
print(merged_df)
# Fill NaN values in goals_df2 with 0
merged_df['goals_df2'].fillna(0, inplace=True)

# Subtract the goals
merged_df['goals_diff'] = merged_df['goals_df1'] - merged_df['goals_df2']
print(merged_df)
# Select only the relevant columns
result_df = merged_df[['player', 'goals_diff']]

# Display the result
print(result_df)
```
